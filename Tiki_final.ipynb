```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_url(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    return soup
    

    
def scrape_tiki(url):
    soup = get_url(url)
    products = soup.find_all('div', {"class":"product-item"})
    
    data = []
    for product in products:
        d = {'seller_id':'', 'product_id':'', 'title':'', 'price':'', 'discounts':'', 'url':'', 'img_url':'', 'review':''}       
        
     
        d['seller_id']   = product['data-seller-product-id']
        d['product_id']  = product.a['data-id']
        d['title'] = product.find('p', {'class':'title'}).text.replace('\n','').strip(' ').split('-')[0].strip()
        d['price'] = product['data-price']
        try:
          d['discounts'] = product.find('p', {'class':'price-sale'}).text.split()[1]
        except:
          pass
        d['url']     = product.a['href']
        d['img_url'] = product.img['src']
        d['review']  = product.find('p',{'class':'review'}).text

            
        data.append(d)
        
    return data
```

```python
for i in range(5):
    if i == 0:
      data = scrape_tiki('https://tiki.vn/dien-thoai-smartphone/c1795?src=tree')
    else:
      a = scrape_tiki('https://tiki.vn/dien-thoai-smartphone/c1795?src=tree&page='+str(i))  
      data = data + a
    
```
